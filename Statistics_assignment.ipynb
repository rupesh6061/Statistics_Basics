**1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales.**
- **Qualitative Data:** It represents category or characteristics This data does not involve numerical data with arithmetic operations. eg of qualitative data are eye color,Gender and marital status.
- **Types of qualitative data:**
- **Nominal Data:** No order in data. e.g. Gender, Race, Eye color,Blood group
- **Ordinal Data:** Categories with meaningful order but difference is not measurable.e.g.  Educational Degree, officers Rank.
- **Quantitative Data:** It consists of numerical data that can be measured or counted. e.g. Height, temperature,Marks obtained in exams.
- **Types of quantitative data:**
- **Interval Data:** Ordered data with equal data but no true zero point. e.g. length, temperature,Scores.
- **Ratio Data:** Ordered data with equal intervals and true zero point. e.g.  Height,Weight and speed of vehicle.  

**2.  What are the measures of central tendency, and when should you use each? Discuss the mean, median and mode with examples and situations where each is appropriate.**
- Measure of central tendency is a statistical tool used for summarizing the dataset by identifying the central value.The three measures are Mean,Median and Mode.Each measure is suitable for different types of data and situations.
- **Mean** The sum of all the values divided by total number of values. e.g. 1,2,3,4,5   mean = 15/5 = 3
- **Use Case**  Used when data is quantitative, continuous and there is no outliers present.
- **Median** Median is the middle value when data is arranged in ascending order.If N is odd then middle value if N is even then average of two middle values.
- e.g. when N is odd 1,4,5,3,2 Values arranged in ascending order = 1,2,3,4,5 , Median = 3
- e.g. when N is even 1,4,6,5,3,2 Values arranged in ascending order = 1,2,3,4,5,6 , Median = 7/2 = 3.5
- **Use Case** When data has outliers or skewed distribution.
- **Mode** The value that appears most often in dataset. e.g.  1,2,2,2,3,3,4,5,5,5,5 Mode = 5 (Since 5 appears most often in dataset)
- **Use Case** Used when data is categorical or when we need to identify most common occurence of value.

**4. What is a box plot, and what can it tell you about the distribution of data?**
- A box plot (box and whisker plot) is a graphical representation of the distribution of a dataset using a five number summary.
- **Minimum** Smallest Observation
- **Q1** 25th percentile
- **Median Q2** 50th percentile
- **Q3** 75th percentile
- **Maximum** Largest observation
- **Box Plot reveals following things about Data Distribution**
- **1. Central Tendency & spread**
- **Median Q2** It shows Data's Center
- **Box (IQR)** The Inter Quartile Range (Q3 - Q1) represents the middle 50% of data.A wider box is equal to more spread.
- **2. Skewness**
- **Right Skewed** Whisker on the right is longer means median leans left.
- **Left Skewed** Whisker on the left is longer means median leans right.
- **Symmetric** Median centered, whiskers equal.
- **3. Variability**
- **Whiskers** It shows range of typical data (1.5 * IQR)
- **Outliers** Points beyond whiskers indicate extreme values.

**5. Discuss the role of random sampling in making inferences about populations?**
- **Estimates Population Parameters** A random sample helps estimates mean or variance for the whole population.eg A survey on 500 voters can predict election results for an entire country. 
- **Allows Hypothesis Testing** We can test claims about the population using sample data. eg testing if a new drug is effective.
- **Supports Confidence Intervals** Random sampling allows us to construct confidence intervals, giving a range in which the true population parameter is likely to fall.
- **Enables Predictive modeling** In machine learning and data analysis, random samples help to create models that generalise well to unseen data.

**6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?** 
- Skewness is a measure of the asymmetry of a data distribution.It tells us whether the data is symmetrically distributed or leans towards one side.
- A symmetrical distribution has zero skewness.
- An asymmetrical distribution has positive or negative skewness.
- **Types of Skewness**
- **Symmetrical(zero Skewness)** The data is evenly distributed around the center.The mean,median and mode are equal. eg heights of the people in a large population.
- **Positive Skewness (Right Skewed)** The right tail is longer than the left. Mean > Median > Mode. eg Income Distribution (Few very high incomes raise the mean).
- **Negative Skewness(Left Skewed)** The left tail is longer than the right.Mean < Median < Mode.eg Age of retirement.
- **Impact of Skewness of Data Interpretation**
- **Effect on Central Tendency Measures**
- Right-skewed data-> Mean is higher than the median
- Left- skewed data-> Mean is lower than the median
- **Affects Statistical Assumptions**
- Many statistical tests assumes zero skewness.
- if data is skewed, transformations (e.g. log transformation) may be needed.
- **Impacts Decision-Making**
- In business, right-skewed income data suggests a few very high earners.
- In healthcare, left-skewed data on hospital stays suggests most people recover quickly, with few staying longer.
- eg: 3,5,7,9,20
- Mean = 8.8, Median = 7 -> Right-skewed

**7. What is the interquartile range (IQR), and how is it used to detect outliers?**
- The IQR measures the spread of the middle 50% of data in a dataset.
- IQR = Q3 - Q1 
- Q1 (First Quartile):25th percentile
- Q3 (Third Quartile) : 75th percentile
- Outliers are extreme values that deviate significantly from other observations.The IQR method identifies the using:
- Lower Bound = Q1 - 1.5 * IQR
- Upper Bound = Q2 + 1.5 * IQR (Data points outside these bounds are considered outliers)
- Example: 2,5,7,10,15,18,22,30,35,50
- Q1 = 7, Q3 = 30
- IQR = Q3 - Q1 = 30-7 = 23
- Lower Bound = 7 -(1.5 * 23) = -27.5
- Upper Bound = 30 + (1.5 * 23) = 64.5
- Any number below -27.5 and above 64.5 is an outlier.

**8.  Discuss the conditions under which the binomial distribution is used.**
- The binomial distribution is a discrete probability used to model the number of successes in a fixed number of independent trials, each with the same probability of success.It applies under following conditions:
- **Fixed Number of Trials** The experiment consists of a fixed number of trials. eg Tossing a coin 10 times
- **Independent Trials** Each trial must be independent i.e. outcome of one trial does not affect another. e.g.  Coin flips are independent.
- **Binary Outcomes** Each trial has only two possible outcomes. i.e. Success and Failure e.g. Medical test result: Positive or Negative
- **Constant Probability of Success(p)**
- The probability of success must remain the same for all trials. e.g. A fair coin always has p = 0.5 for heads.

**9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).**
- The normal distribution (also called the Gaussian distribution) is a bell shaped probability distribution where most data points cluster around the mean and values further from the mean become less frequent. e.g. Heights of people and IQ scores
- **Properties of the Normal Distribution**
- **Bell-shaped Curve** Symmetrical around the mean
- **Mean = Median = Mode** Highest point of the curve
- **Symmetry** Left and right halves are mirror images
- **Asymptotic** The curve never touches the x-axis
- **Defined by Two Parameters**
- **Mean** Center of the distribution
- **Standard Deviation** spread of the data
- The **Empirical Rule**  describes how data is distributed in a normal distribution based on standard deviations from the mean.
-  68% of data falls within 1 standard deviation of the mean.
-  95% of data falls within 2 standard deviations of the mean.
-  99.7% of data falls within 3 standard deviations of the mean.

**10. Provide a real-life example of a Poisson process and calculate the probability for a specific event**
- A poisson process models events that occur randomly and independently over time or space, with a known average rate (𝜆).
- Example:  A bank receives an average of 10 customers per hour (𝜆 = 10). We want to find the probability that exactly 8 customers arrive in an hour. 

```python
import math

customers_per_hour = 10 
k = 8 
poisson = (math.exp(-customers_per_hour)*(customers_per_hour**k))/ math.factorial(k)
poisson
# The probability of exactly 8 customers arriving at the bank in one hour is 0.1126 or 11.26%
```

**11. Explain what a random variable is and differentiate between discrete and continuous random variables.**
- It is a set of possible values from a random experiment.It represents the possible outcomes of a probabilistic event.e.g. Tossing a coin , outcomes can be represented by a random variable:
- X = 1 for Heads
- X = 0 for Tails
- **Types of Random Variable**
- **1. Discrete Random Variable** A discrete random variable takes on a finite or countable number of values.
- It often used in binomial and poisson distributions.
- It is used when outcomes are countable. e.g. number of calls in a day.
- **2. Continuous Random Variable** A continuous random variable takes infinite values within a given range.
- It often follows normal distribution.
- Continuous random variables are used when outcomes are measurable. e.g. weight,height,time

```python
#12. Provide an example dataset, calculate both covariance and correlation, and interpret the results.
import numpy as np 
import pandas as pd 
data = {
    'Study_hours': [2,3,4,5,6,7,8],
    'Exam_Score' : [50,55,60,65,60,75,80]
}
df = pd.DataFrame(data) 
covariance = df.cov().loc['Study_hours', 'Exam_Score']
correlation = df.corr().loc['Study_hours', 'Exam_Score']
covariance, correlation
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Study_hours': [2, 3, 4, 5, 6, 7, 8],
    'Exam_Score':  [50, 55, 60, 65, 70, 75, 80]
}

df = pd.DataFrame(data)
plt.figure(figsize=(8,5))
sns.regplot(x=df['Study_hours'], y=df['Exam_Score'], scatter_kws={'color' : 'blue'}, line_kws={'color':'red'})
plt.title("Scatter Plot with Regression Line")
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.grid()
plt.show()
```

```python
cov_matrix = df.cov()
corr_matrix = df.corr()


fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.heatmap(cov_matrix, annot=True, cmap="coolwarm", fmt=".2f", ax=axes[0])
axes[0].set_title("Covariance Matrix")

sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", ax=axes[1])
axes[1].set_title("Correlation Matrix")

plt.show()
```
